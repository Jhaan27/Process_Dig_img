```python
from google.colab import drive
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Monta o Google Drive
drive.mount('/content/drive')

# Caminho para a pasta no Google Drive onde suas imagens estão armazenadas
folder_path = '/content/drive/MyDrive/'

# Lista os arquivos na pasta
files_list = os.listdir(folder_path)
print("Arquivos na pasta:", files_list)

# Escolhe uma imagem para carregar (substitua 'nome_da_imagem.jpg' pelo nome real do arquivo)
chosen_file = 'peb.png'

# Caminho completo para a imagem escolhida
image_path = os.path.join(folder_path, chosen_file)

# Lê a imagem usando OpenCV (certifique-se de que é uma imagem em tons de cinza)
imagex = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Exibe a imagem
plt.imshow(imagex, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Carregando a imagem
img = imagex.copy()

# Aplicando um limiar para binarizar a imagem (se necessário)
_, thresh = cv2.threshold(img, 135, 255, cv2.THRESH_BINARY)

# Definindo um kernel para as operações morfológicas
kernel1 = np.ones((3, 3), np.uint8)
kernel2 = np.ones((5, 5), np.uint8)

# Aplicando a operação de abertura para remover ruído
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel1,iterations=1)
# Aplicando dilatação seguida por erosão (fechamento)
closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel2, iterations=2)

closing = cv2.morphologyEx(closing, cv2.MORPH_DILATE, kernel2, iterations=1)


# Plota as imagens lado a lado no domínio do tempo
plt.figure(figsize=(20, 15))

# Exibindo a imagem original, a imagem binarizada e a imagem após a operação de abertura
plt.subplot(131), plt.imshow(img, 'gray'), plt.title('Imagem Original')
#plt.subplot(132), plt.imshow(thresh, 'gray'), plt.title('Imagem Binarizada')
plt.subplot(132), plt.imshow(opening, 'gray'), plt.title('Imagem após Erosão')
plt.subplot(133), plt.imshow(closing, 'gray'), plt.title('Imagem após Dilatação')
plt.show()
```
